data with min, max, average and std of 'Profit' & 'Sales' for each Sub-Category under each Category

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('global_superstore_2016.csv', encoding=('ISO-8859-1'))
```

```python
df['Profit'] = df['Profit'].str.strip("(")
df['Profit'] = df['Profit'].str.strip("$")
df['Profit'] = df['Profit'].str.strip(")")
```

```python
df['Profit'] = df['Profit'].str.replace(',', '')
df['Profit'] = df['Profit'].astype(float)
```

```python
df['Sales'] = df['Sales'].str.strip("$")
```

```python
df['Sales'] = df['Sales'].str.replace(',', '')
```

```python
df['Sales'] = df['Sales'].astype(float)
```

```python
alll = df.groupby(['Category','Sub-Category'],as_index=True).Profit.max()
```

```python
alll
```

```python
alll2 = df.groupby(['Category','Sub-Category'],as_index=True).Profit.min()
```

```python
alll2
```

```python
alll3 = df.groupby(['Category','Sub-Category'],as_index=True).Profit.mean()
alll3
```

```python
sale = df.groupby(['Category','Sub-Category'],as_index=True).Sales.max()
```

```python
sale
```

```python
sale2 = df.groupby(['Category','Sub-Category'],as_index=True).Sales.min()
sale2
```
